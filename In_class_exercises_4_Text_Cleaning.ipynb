```python
# Setup (run this cell first)
import re
import pandas as pd
```

## Q1 (1 point) — Answer below

```python
# Q1 — ANSWER CELL
FILE_PATH = "support_messages.txt"

# TODO: load the TAB-separated file into df
# Hint: pd.read_csv(FILE_PATH, sep="\t")
df = df = pd.read_csv(FILE_PATH, sep="\t", names=["id", "message"])

# TODO: print df.shape and df.head(3)
print("Shape:", df.shape)
print("\nFirst 3 rows:")
print(df.head(3))
```

## Q2 (3 points) — Answer below

```python
# Q2 — ANSWER CELL
# TODO: create word_count, char_count, num_count, upper_word_count
# Hint for digits: df["message"].str.count(r"\d")
# Hint for ALL-CAPS words: count tokens where token.isupper()
import re

# word_count
df["word_count"] = df["message"].astype(str).apply(lambda x: len(x.split()))

# char_count
df["char_count"] = df["message"].astype(str).apply(len)

# num_count (digits 0–9)
df["num_count"] = df["message"].astype(str).str.count(r"\d")

# upper_word_count (ALL-CAPS words)
df["upper_word_count"] = df["message"].astype(str).apply(
    lambda x: len([token for token in x.split() if token.isupper()])
)

# Display full DataFrame
print(df)

# TODO: display/print the full DataFrame
```

## Q3 (3 points) — Answer below

```python
# Q3 — ANSWER CELL
# Option A (sklearn stopwords):

# from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
# STOPWORDS = set(ENGLISH_STOP_WORDS)
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

STOPWORDS = set(ENGLISH_STOP_WORDS)

def clean_text(text: str) -> str:
    text = str(text)

    # 1️⃣ lowercase
    text = text.lower()

    # 2️⃣ remove punctuation/symbols (keep letters/numbers/spaces)
    text = re.sub(r"[^a-z0-9\s]", "", text)

    # 3️⃣ remove stopwords
    words = text.split()
    words = [word for word in words if word not in STOPWORDS]

    # 4️⃣ remove extra spaces
    text = " ".join(words)

    return text


# Ensure id column is numeric
df["id"] = pd.to_numeric(df["id"], errors="coerce")

# Create clean column
df["clean"] = df["message"].apply(clean_text)

# Correct filtering
print(df.loc[df["id"].isin([1, 4]), ["id", "message", "clean"]])
```

## Q4 (2 points) — Answer below

```python
# Q4 — ANSWER CELL
# order_id pattern: r"ORD-\d{4}" with re.IGNORECASE
# email pattern (simple): r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"

# TODO: create df["order_id"] and df["email"]
# TODO: print/display df[["id", "order_id", "email"]]
import re

# Extract order_id (case-insensitive)
df["order_id"] = df["message"].str.extract(r"(ORD-\d{4})", flags=re.IGNORECASE)

# Extract email
df["email"] = df["message"].str.extract(
    r"([A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,})"
)

# Print required columns
print(df[["id", "order_id", "email"]])
```

## Q5 (1 point) — Answer below

```python
# Q5 — ANSWER CELL
# Hint: from sklearn.feature_extraction.text import TfidfVectorizer
# 1) fit TF-IDF on df["clean"]
# 2) compute average TF-IDF per term across documents
# 3) print top 5 terms + their average scores

# TODO
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Fit TF-IDF on clean column
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df["clean"])

# Get feature names
feature_names = np.array(vectorizer.get_feature_names_out())

# Compute average TF-IDF score per term
avg_tfidf = tfidf_matrix.mean(axis=0).A1

# Get top 5 terms
top_indices = avg_tfidf.argsort()[-5:][::-1]

print("Top 5 Keywords with Highest Average TF-IDF:")
for idx in top_indices:
    print(feature_names[idx], ":", round(avg_tfidf[idx], 4))
```
